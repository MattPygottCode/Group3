```python
import ast

smallest_factor_data = [1, 2, 3, 4, 5]

with open("file1.txt", "w") as file:
    file.write(str(smallest_factor_data))

with open("file1.txt", "r") as file:    
    content = file.read()

restored_list = ast.literal_eval(content)

print(restored_list)
```

```python
import json

a = [1, 2, 3]
b = [10, 20, 30]
c = [7, 8, 9]

data = {
    "a": a,
    "b": b,
    "c": c
}

with open("lists.json", "w") as f:
    json.dump(data, f)

with open("lists.json", "r") as f:
    data = json.load(f)

a = data["a"]
b = data["b"]
c = data["c"]

print(a, b, c)
```
